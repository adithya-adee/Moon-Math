```python
# Exercise 14

problems = [[5,19], [-4,9], [13,0], [0,0] ]
n = 13

for (a,b) in problems:
    rem1 = a % n
    rem2 = b % n

    if rem1 == rem2:
        print(f"{a} ≅ {b} mod {n}")
    else:
        print(f"({a},{b}) mod {n} is incongruent")
```

```python
# Exercise 15 , 16, 17, 18

def find_congruence_solutions():
    # x = 4 (mod 6)
    expected_remainder = 4 % 6
    x = 0
    solutions = []

    print(expected_remainder)

    # x = 6k + 4
    while x < 25:
        current_remainder = x % 6
        
        if current_remainder == expected_remainder:
            solutions.append(x)
        
        x += 1

    print(solutions)

find_congruence_solutions()
```
